###Let's look at what happens when we multiply the Bark transform with the cosine matrix used in the filter design process. 
###For analysis purposes, let's compare these steps for both types of Bark domain transforms (RASTA and PEAQ construction methods).

###1. RASTA

```python
nfft=2048
fs=48000
nfilts=109
version="rasta"
Br = Bark(nfft, fs, nfilts, version)
```

```python
L = nfft/2
M = 64  # Number of filter coefficients

w = (pi/L)*(np.arange(L)).T  # Frequency grid

# Rectangular C matrix
w_rs = w.reshape(w.shape[0],1)
M_rs = np.arange(M)
M_rs = M_rs.reshape(M_rs.shape[0],1)
C = cos(np.dot(w_rs,M_rs.T))
```

```python
print C.shape
```

```python
# Matrix product of Bark transform (B.W) and cosine matrix (C)
BCr = np.dot(Br.W[:,0:Br.nfft/2],C)

print BCr.shape
```

```python
fig = plt.figure(0, (8,8))

ax = fig.add_subplot(111)
ax.imshow(log(BCr))
ax.set_aspect('auto')
```

```python
fig = figure(0, (16,12))

for i in range(0,25):
    subplot(5,5,i+1)
    plot(C[20*i,:])
    grid(1)
```

```python
BWr = Br.W[:,0:Br.nfft/2]
BWr.shape
```

```python
imshow(BWr)
```

```python
plot(BWr[80,:])
grid(1)
```

```python
fig = figure(0, (16,12))

for i in range(0,25):
    subplot(5,5,i+1)
    plot(BCr[4*i,:])
    grid(1)
```

These are very poor results. There appears to be cancellations occurring in these, especially as frequency increases. The RASTA method appears to have some issues.

###2. PEAQ

```python
nfft=2048
fs=48000
nfilts=109
version="peaq"
Bp = Bark(nfft, fs, nfilts, version)
```

```python
# Matrix product of Bark transform (B.W) and cosine matrix (C)
BCp = np.dot(Bp.W[:,0:Bp.nfft/2],C)

print BCp.shape
```

```python
fig = plt.figure(0, (8,8))

ax = fig.add_subplot(111)
ax.imshow(log(BCp))
ax.set_aspect('auto')
```

```python
BWp = Bp.W[:,0:Bp.nfft/2]
BWp.shape
```

```python
imshow(BWp)
```

```python
plot(BWp[80,:])
grid(1)
```

```python
fig = figure(0, (16,12))

for i in range(0,25):
    subplot(5,5,i+1)
    plot(BCp[4*i,:])
    grid(1)
```

These results appear to be much more promising than the results from the RASTA Bark transform! The cancellations are not occurring here. Each product result appears to take full shape.

Since this approach appears to be working more smoothly, let's redo the filter design process using the PEAQ Bark transform.

####Rename variables

```python
B = Bp
BC = BCp
```

####Matrix BC is not square, so it needs to be full rank in order for us to be able to compute its pseudo-inverse. Let's check its rank.

```python
BC_rank = np.linalg.matrix_rank(BC)

print 'Shape of BC: ', BC.shape
print 'Rank of BC: ', BC_rank
```

Letting k = rank(BC) and BC have dimension m x n, we see that k < m and k < n. So BC is not full rank here, and thus is rank deficient.

####Compute the SVD and the pseudo-inverse of matrix BC using the SVD approach in Numpy

```python
U_BC, S_BC, V_T_BC = np.linalg.svd(BC)

BC_pinv = np.linalg.pinv(BC)
```

```python
# Plot the singular values of BC (decreasing order)
plot(S_BC)
```

```python
# Compute the matrix product of the inverse and the original matrix to see
# how close they are to the identity matrix.
I_BC_papprox = np.dot(BC_pinv, BC)

fig = plt.figure(0, (8,8))

ax = fig.add_subplot(111)
ax.imshow(I_BC_papprox)
ax.set_aspect('auto')
```

Not bad! This is much closer to the identity matrix than what was being produced before using the non-SVD pseudo-inverse approach. But how close is it? One thing we could look at is the square error between the approximate identity matrices and the actual identity matrix.

####Compute the square error between the approximate identity matrices and the actual identity matrix.

```python
diff_BC = I_BC_papprox - np.eye(I_BC_papprox.shape[0])

sq_error_BC = np.mean(np.power(diff_BC,2))

print 'Square Error (BC): ', sq_error_BC
```

The square error from BC is now much smaller! This appears to be working the way we would like for it to.

####Design the filter using the direct linear algebra steps: $\textbf{a} \approx (BC)^{+} \textbf{BD}$

```python
a = np.dot(BC_pinv, BD)
```

```python
# Form impulse response h(n)
h1 = a[M:0:-1]
h2 = 2*a[0]
#h2 = h2.reshape(h2.shape[0],1)
h3 = a[1:M+1]

h12 = np.concatenate((h1,[h2]))

h = np.concatenate((h12,h3))/2

h = np.divide(h,np.amax(np.abs(h)))
```

One of the reasons working with music signals can be challenging is the inherit non-stationarity of musical sounds, especially acoustic instruments. A high level example of this non-stationarity is the inconsistent system behavior between transients and sustains. 

Here we'll make an effort seperate transients and sustains using wavelet decomposition. In practice, the statistical nonstationarity breaks down over more than just transient/sustain behavior, this represents a first attempt at reducing non-stationarity using a transient/sustain paradigm.

##Compute Wavelet Coefficients

```python
import pywt

#Compute SWT
def computeWaveletCoefficients(data, name='db4', numLevels = 2):
    #Create wavelet object
    w = pywt.Wavelet(name)
    
    #Compute wavelet coefficients, use swt to avoid decimation:
    c = pywt.swt(data, w, level=numLevels)
    
    #Put all coefficients into a big matrix, XFF
    XFF = np.zeros((len(data), numLevels + 1))

    XFF[:,0] = c[0][0]
    for i in range(len(c)):
        XFF[:,i+1] = c[i][1]
        
    #Throw out last 1000 samples, wierd stuff tends to happen at the wend of wavelet 
    return XFF[:-1000,:]
```

```python
mic1 = alignedSignals['mic1']
piezo = alignedSignals['piezo']
```

```python
c = computeWaveletCoefficients(piezo, name='db20', numLevels=4)

#Normalize Wavelet Coefficients:
normalizationCoeffients = amax(abs(c),axis=0)
cNorm = abs(c)/normalizationCoeffients
```

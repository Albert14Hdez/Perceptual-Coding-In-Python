We previously explored the Frequency-Domain Least Squares filter design method, and we found this to provide a potentially useful way of modeling responses as well as comparing two frequency-domain signals. 

If this method provides a way to design filters from the frequency domain, is it also possible to design filters from the Bark domain? The Bark domain would provide some advantages for the filter design process; specifically, it would allow for the measurement of perceptual, psychoacoustical differences instead of simply magnitude differences across DFT bins.

In this notebook, I will begin to explore the process of bringing this frequency-domain LS process into the Bark domain via the Bark transform.

##Mathematical Exploration

Let's revisit some of the important formulas used in the frequency-domain LS formulation and derivations:

* Define $A(\omega)$ as the actual amplitude response, $D(\omega)$ as the ideal amplitude response, $W(\omega)$ as a non-negative weighting function.
* Given a filter length N, a desired amplitude function $D(\omega)$, and a non-negative weighting function $W(\omega)$, find the symmetric filter that minimizes the weighted integral square error (or "L2 error"), defined by:

$$ \varepsilon_2 = \int_{0}^{\pi} W(\omega)[A(\omega) - D(\omega)]^2 d\omega$$ 

* Once the filter length N and filter Type (I, II, III, or IV) are chosen, the goal is to find the filter coefficients h(n) that minimize $\varepsilon_2$.
* Here, we consider Type I FIR filters. Type I = Symmetric Impulse Response with Odd Length
* For Type I filters, the amplitude response is given by:

$$A(\omega) = \sum_{n=0}^{M} a(n) cos(\omega n)$$

where $$a(0) = h(M)$$

$$a(n) = 2h(M-n), for 1 \leq n \leq M$$

$$M = \frac{N-1}{2}$$

###General Discrete Square Error

When no analytic expressions are available for the inverse DTFT of $D(\omega)$ and $W(\omega)$, then one may consider the discrete square error measure:

$$\varepsilon_2 = \sum_{k=1}^{L} W(\omega_k) (A(\omega_k) - D(\omega_k))^2$$

where $\omega_k$ are a set of frequency points which are not necessarily uniformly spaced.

Now, define 

$$A_k = A(\omega_k)$$

$$D_k = D(\omega_k)$$

$$W_k = W(\omega_k), 1 \leq k \leq L$$

Then the error function

$$e_k = A_k - D_k$$

or 

$$e_k = \sum_{n=0}^{M} a(n)\cos(n\omega_k) - D_k$$

can be written in matrix form as:

$$\begin{bmatrix}
    e_1  \\
    e_2  \\
    ...     \\
    e_L  
\end{bmatrix} = 
\begin{bmatrix}
    1      & cos(\omega_1) & \dots & cos(M\omega_1) \\
    1      & cos(\omega_2) & \dots & cos(M\omega_2) \\
    ...       & ... & ... & ... \\
    1      & cos(\omega_L) & \dots & cos(M\omega_L)
\end{bmatrix} 
\begin{bmatrix}
    a(0)  \\
    a(1)  \\
    ...     \\
    a(M)  
\end{bmatrix}
-
\begin{bmatrix}
    D_0  \\
    D_1  \\
    ...     \\
    D_L  
\end{bmatrix}$$

or, more briefly, as:

$$\textbf{e} = \textbf{Ca} - \textbf{D}$$

Matrix C has more rows than columns, and we can see that $\textbf{Ca} = \textbf{D}$ is an over-determined system of equations.

The least squares solution to an over-determined system minimizes the (unweighted) sum of squares of the error vector, which can be compactly written as:

$$\varepsilon_2 = e^{T} e$$

The least squares solution is obtained by first multiplying both sides of $\textbf{Ca} \approx \textbf{D}$ on the left by $C^{T}$ to obtain the "normal equations"

$$C^{T}\textbf{Ca} = C^{T}\textbf{D}$$

The least squares (unweighted) solution is then:

$$\textbf{a} = (C^{T}C)^{-1} C^{T} D$$

The weighted square error solution is obtained by first multiplying on the left by $W = diag(\sqrt(W_k))$:

$$\textbf{WCa} \approx \textbf{WD}$$

and then by $C^{T}$:

$$C^{T} WCa = C^{T} WD$$

Solving this equation gives:

$$\textbf{a} = (C^{T}WC)^{-1} C^{T}WD$$

Using this solution, the coefficients $a(n)$ can be computed based on a specified filter length N, a desired ideal response $D(\omega_k)$, and a specified non-negative weighting function $W(\omega_k)$. From coefficients $a(n)$, the corresponding filter coefficients $h(n)$ can easily be computed.

##Transform to Bark Domain

Now, is it possible to take the steps from above and incorporate the Bark transform into the formulation, to make the filter design process actually take place in the Bark domain? If so, what happens in the Bark domain? Can we produce interesting filters? Can we produce results at all?! Let's explore and find out.

* The Bark transform is represented by a matrix, $B$. The dimension of $B$ is **nfilts x (nfft/2)**, where **nfilts** is the number of filters or critical bands used in the mapping to the Bark scale, and **nfft** is the FFT size being used. The transform maps from the Fourier (FFT) frequency domain to the Bark domain.
* There is also an inverse Bark transform, $B_{inv}$ that maps back to the Fourier domain. The dimension of $B_{inv}$ is **M x M**, where **M** is the length of the filter $a(n)$ that is being designed.

Let's look at the non-weighted least squares problem from above:

$$\textbf{e} = \textbf{Ca} - \textbf{D}$$

or, in solution form:

$$\textbf{Ca} \approx \textbf{D}$$

The thought here is that we could multiply both sides of $\textbf{Ca} \approx \textbf{D}$ by $\textbf{B}$ to map the problem to the Bark domain:

$$\textbf{BCa} \approx \textbf{BD}$$

Vector $\textbf{D}$ is already in the frequency domain, so $\textbf{BD}$ maps directly to Bark domain.
Vector $\textbf{a}$ is a "time-domain" sequence.
Matrix $\textbf{C}$ is a cosine matrix that equates to a simplified DFT matrix for the case of a Type I filter (symmetric impulse response with odd length), so the matrix product $\textbf{BC}$ is some form of a combined simplified DFT-Bark transform.

Following the approach above, both sides of the equation can be multiplied on the left by $(BC)^{T}$:

$$(BC)^{T}\textbf{BCa} = (BC)^{T}\textbf{BD}$$

The least squares (unweighted) solution is then:

$$\textbf{a} = ((BC)^{T}BC)^{-1} (BC)^{T} \textbf{BD}$$

Below, I will explore this formulation to see what comes of it. What is the form of the sequence $\textbf{a(n)}$ that we get out of this computational process? Are the assumptions we made about this filter applicable here? Let's see what we discover.

###Let's look at what happens when we multiply the Bark transform with the cosine matrix used in the filter design process

```python
L = nfft/2
M = 64  # Number of filter coefficients

w = (pi/L)*(np.arange(L)).T  # Frequency grid

# Rectangular C matrix
w_rs = w.reshape(w.shape[0],1)
M_rs = np.arange(M)
M_rs = M_rs.reshape(M_rs.shape[0],1)
C = cos(np.dot(w_rs,M_rs.T))
```

```python
print C.shape
```

```python
# Matrix product of Bark transform (B.W) and cosine matrix (C)
BC = np.dot(B.W[:,0:B.nfft/2],C)

print BC.shape
```

```python
fig = plt.figure(0, (8,8))

ax = fig.add_subplot(111)
ax.imshow(BC)
ax.set_aspect('auto')
```

```python
fig = plt.figure(0, (16,8))

ax = fig.add_subplot(121)
ax.imshow(C)
ax.set_aspect('auto')
ax = fig.add_subplot(122)
ax.imshow(B.W[:,0:B.nfft/2])
ax.set_aspect('auto')
```

####Compute $(BC)^{T}BC$ and $( (BC)^{T}BC )^{-1}$

```python
BC_T_BC = np.dot(BC.T,BC)
BC_T_BC_inv = np.linalg.inv(BC_T_BC)
#a1 = np.linalg.inv(np.dot(BC.T,BC))
#a2 = np.dot(a1,BC.T)
```

```python
print BC_T_BC.shape
print BC_T_BC_inv.shape
```

```python
fig = plt.figure(0, (16,8))

ax = fig.add_subplot(121)
ax.imshow(BC_T_BC)
ax.set_aspect('auto')
ax = fig.add_subplot(122)
ax.imshow(BC_T_BC_inv)
ax.set_aspect('auto')
```

```python
I_hat = np.dot(BC_T_BC_inv,BC_T_BC)
```

```python
print I_hat.shape
print I_hat
```

```python
diff = I_hat - np.eye(I_hat.shape[0])
```

```python
sq_error = np.mean(np.power(diff,2))
print sq_error
```

```python
fig = plt.figure(0, (8,8))

ax = fig.add_subplot(111)
ax.imshow(I_hat)
ax.set_aspect('auto')
```

####Check the rank of $BC$ and $(BC)^{T}BC$

```python
BC_rank = np.linalg.matrix_rank(BC)

BC_T_BC_rank = np.linalg.matrix_rank(BC_T_BC)

print 'Shape of BC: ', BC.shape
print 'Rank of BC: ', BC_rank
print 'Shape of BC_T_BC: ', BC_T_BC.shape
print 'Rank of BC_T_BC: ', BC_T_BC_rank
```

Matrix $BC$ is rank deficient! (i.e. matrix $BC$ of size **m x n** (here, m = 109, n = 64), the rank of $BS$ is less than both m and n).

This is not good for us. Since $BC$ is both non-square and rank deficient, it has no inverse and the Moore-Penrose pseudoinverse cannot be computed for it.

Let's see what happens in the next steps. We should see some very poor results in the end...

###Create the desired response for the filter design step

```python
# Parameters
L_f = 2048  # Frame length
L_s = len(mic1)  # Signal length

# Piezo signal DFT
#P = np.fft.fft(piezo[0:L_f],nfft)
P = np.fft.fft(piezo,nfft)
P_abs = abs(P)
P_sub = P_abs[0:nfft/2]

# Mic signal DFT
#M = np.fft.fft(mic1[0:L_f],nfft)
Mic = np.fft.fft(mic1,nfft)
Mic_abs = abs(Mic)
Mic_sub = Mic_abs[0:nfft/2]
```

```python
# Approximate the "transfer function" between the microphone and piezo signals
F = Mic_sub/P_sub

# The desired response is this "transfer function"
D = F
```

```python
plot(D)
```

###Create the needed pieces and attempt to design the filter

```python
BD = np.dot(B.W[:,0:B.nfft/2],D)
```

```python
plot(BD)
```

###Design the filter using the direct linear algebra steps

```python
a1 = np.dot(BC_T_BC_inv, BC.T)
a = np.dot(a1, BD)
```

```python
plot(a)
```

```python
# Form impulse response h(n)
h1 = a[M:0:-1]
h2 = 2*a[0]
#h2 = h2.reshape(h2.shape[0],1)
h3 = a[1:M+1]

h12 = np.concatenate((h1,[h2]))

h = np.concatenate((h12,h3))/2

h = np.divide(h,np.amax(np.abs(h)))
```

```python
plot(h)
```

```python
H = np.fft.fft(h,nfft)
H_abs = abs(H)

plot(H_abs[0:nfft/2])
```

```python
from scipy.signal import lfilter

mic1_approx = lfilter(h,1,piezo)

mic1_approx = mic1_approx/max(abs(mic1_approx))
```

```python
plot(mic1_approx)
```

###Design the filter using Least Squares

```python
# Solve linear system to get a(n)
a = np.linalg.lstsq(BC, BD)
a = a[0]
```
